```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

pclass , sex ,sibsp , parch , embarked , class , who , adult_male , embark_town , alive , alone

```python
import pandas as pd

dfx = pd.get_dummies(df['sex'])
dfx
```

```python
df = pd.concat([df,dfx], axis = 1) # df 와 dfx 를 x축으로 나란하게 union 시킴 . 단, size가 동일해야 함
df
```

```python
dfx = pd.get_dummies(df['pclass'],prefix='pclass')
df = pd.concat([df,dfx],axis=1)
df.head(4)
```

female ,	male , pclass_1	, pclass_2 ,	pclass_3

```python
df[[ 'fare','female','male','pclass_1','pclass_2','pclass_3','survived']].head(5)
```

```python
X = df[[ 'fare','female','male','pclass_1','pclass_2','pclass_3']]
Y = df[[ 'survived']]
X.shape , Y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X) # 값은 가지고 있지 않고 규칙만 가지고 있게 됨
```

```python
# 실제 데이터 넣어서 값을 확인하면됨
X = scaler.transform(X)
X , X.shape
```

```python
# 일부 데이터 확인 하기 위해 split 한다
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(X, Y)
```

```python
# 선형회귀에 실제 대입
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
logR.fit(x_train , y_train)
```

```python
logR.score(x_train,y_train)
```
